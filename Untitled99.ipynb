```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import zipfile as zipfile
import csv
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.image as mpimg
import os
```

```
local_dir = '/content/archive (31).zip'
zip_ref = zipfile.ZipFile(local_dir,'r')
zip_ref.extractall('/tmp/')
zip_ref.close()
```

```
train_dir='/tmp/tomato/train/'
train_bacterial = os.path.join(train_dir,'Tomato___Bacterial_spot')
train_early = os.path.join(train_dir,'Tomato___Early_blight')

train_late = os.path.join(train_dir,'Tomato___Late_blight')
train_leaf = os.path.join(train_dir,'Tomato___Leaf_Mold')
train_septoria = os.path.join(train_dir,'Tomato___Septoria_leaf_spot')
train_spider = os.path.join(train_dir,'Tomato___Spider_mites Two-spotted_spider_mite')
train_target = os.path.join(train_dir,'Tomato___Target_Spot')
train_tomato = os.path.join(train_dir,'Tomato___Tomato_Yellow_Leaf_Curl_Virus')
train_tomatomo = os.path.join(train_dir,'Tomato___Tomato_mosaic_virus')
train_healthy = os.path.join(train_dir,'Tomato___healthy')
```

```
val_dir='/tmp/tomato/val/'
val_bacterial = os.path.join(val_dir,'Tomato___Bacterial_spot')
val_early = os.path.join(val_dir,'Tomato___Early_blight')

val_late = os.path.join(val_dir,'Tomato___Late_blight')
val_leaf = os.path.join(val_dir,'Tomato___Leaf_Mold')
val_septoria = os.path.join(val_dir,'Tomato___Septoria_leaf_spot')
val_spider = os.path.join(val_dir,'Tomato___Spider_mites Two-spotted_spider_mite')
val_target = os.path.join(val_dir,'Tomato___Target_Spot')
val_tomato = os.path.join(val_dir,'Tomato___Tomato_Yellow_Leaf_Curl_Virus')
val_tomatomo = os.path.join(val_dir,'Tomato___Tomato_mosaic_virus')
val_healthy = os.path.join(val_dir,'Tomato___healthy')
```

```
bacterial_files= os.listdir(train_bacterial)
early_files= os.listdir(train_bacterial)
late_files=os.listdir(train_late)
leaf_files= os.listdir(train_leaf)
septoria_files=os.listdir(train_septoria)
spider_files= os.listdir(train_spider)
target_files=os.listdir(train_target)
tomato_files = os.listdir(train_tomato)
tomatomo_files= os.listdir(train_tomatomo)
healthy_files= os.listdir(train_healthy)
```

```
pic_index=2
next_bacterial = [os.path.join(train_bacterial,fname) for fname in bacterial_files[pic_index-2:pic_index]]
next_early=  [os.path.join(train_early,fname) for fname in early_files[pic_index-2:pic_index]]
next_late =  [os.path.join(train_late,fname) for fname in late_files[pic_index-2:pic_index]]
next_leaf=  [os.path.join(train_leaf,fname) for fname in leaf_files[pic_index-2:pic_index]]
next_septoria =  [os.path.join(train_septoria,fname) for fname in septoria_files[pic_index-2:pic_index]]
next_spider=  [os.path.join(train_spider,fname) for fname in spider_files[pic_index-2:pic_index]]
next_target =  [os.path.join(train_target,fname) for fname in target_files[pic_index-2:pic_index]]
next_tomato=  [os.path.join(train_tomato,fname) for fname in tomato_files[pic_index-2:pic_index]]
next_tomatomo = [os.path.join(train_tomatomo,fname) for fname in tomatomo_files[pic_index-2:pic_index]]
next_healthy =  [os.path.join(train_healthy,fname) for fname in healthy_files[pic_index-2:pic_index]]
```

```
train_datagen = ImageDataGenerator(rescale=1.0/255.,fill_mode= 'nearest',rotation_range=40,zoom_range=0.2,shear_range=0.2,horizontal_flip=True,height_shift_range=0.2,width_shift_range=0.2)
```

```
test_datagen = ImageDataGenerator(rescale=1.0/255.)
```

```
train_data_generator = train_datagen.flow_from_directory(train_dir,target_size=(150,150),class_mode='categorical',batch_size=32)
valid_data_generator= test_datagen.flow_from_directory(val_dir,target_size=(150,150),class_mode='categorical',batch_size=32)
```

```
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Conv2D(64,(3,3),input_shape=(150,150,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D((2,2)),
                                    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D((2,2)),
                                    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D((2,2)),
                                    tf.keras.layers.Conv2D(128,(3,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D((2,2)),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dropout(0.9),
                                    tf.keras.layers.Dense(512,activation='relu'),
                                    tf.keras.layers.Dense(10,activation='softmax')
                                    


])
```

```

model.compile(loss='categorical_crossentropy',optimizer= 'rmsprop',metrics=['accuracy'])
history=model.fit(train_data_generator,epochs=20,validation_data= valid_data_generator)
```

```
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss= history.history['loss']
val_loss= history.history['val_loss']
epochs = range(len(acc))
plt.plot(epochs,acc,'r',label='accuracy')
plt.plot(epochs,val_acc,'b',label='val_accuracy')
plt.title('accuracy and val_accuracy')
plt.show()
plt.plot(epochs,loss,'r',label='loss')
plt.plot(epochs,val_loss,'b',label='val_loss')
plt.title('loss and val_loss')
plt.show()
```
